```python
# mjxenv

import os
os.environ["MUJOCO_GL"] = "egl"
import numpy as np
import gymnasium as gym

env = gym.make(
    "Ant-v5",
    render_mode="rgb_array",
    camera_name="egocentric",
    width=64,
    height=64,
    xml_file="/mnt/disk2/xh2/ai4neuron/derl/custom_envs/asset/ant_mjx.xml",
)
obs, info = env.reset()
try:
    print(f"{obs.shape = }")
except AttributeError:
    print(f"{obs = }")
print(f"{env.action_space.shape = }")

action_size = env.action_space.shape[0]

def random_action(action_size):
    return np.random.uniform(-1, 1, (action_size))

obs, reward, terminated, truncated, info = env.step(random_action(action_size))
print(f"{reward = }")
print(f"{terminated = }")
print(f"{truncated = }")
print(f"{info = }")

from tqdm import tqdm
import time
import mediapy as media

repeat_times = 2_000
st = time.time()
video1 = []
for i in tqdm(range(repeat_times)):
    o, r, d, t, info = env.step(random_action(action_size))
    rgb_image = env.render()
    video1.append(rgb_image)
    
et = time.time()

print(f"fps: {repeat_times / (et - st)}")
# show video
media.show_video(video1, fps=30)
```

```python
import dmcenv
import numpy as np
from tqdm import tqdm
import time
import mediapy as media

def visualize_env(env):
    env.debug(True)
    obs, info = env.reset()
    # print(f"{obs = }")
    print(f"{env.action_space.shape = }")
    action_size = env.action_space.shape[0]
    obs, reward, terminated, truncated, info = env.step(random_action(action_size))
    print(f"{reward = }")
    print(f"{terminated = }")
    print(f"{truncated = }")
    print(f"{info = }")
    repeat_times = 1_000
    st = time.time()
    video1 = []
    video2 = []
    for i in tqdm(range(repeat_times)):
        obs, reward, terminated, truncated, info = env.step(random_action(action_size))
        if i % 100 == 0:
            obs, info = env.reset()
        video1.append(obs["image"])
        video2.append(env._env.physics.render(camera_id=0, width=128, height=128))
    et = time.time()

    print(f"fps: {repeat_times / (et - st)}")

    media.show_video(video1, fps=30)
    media.show_video(video2, fps=30)

visualize_env(dmcenv.DMCEnv(walker_name="Ant", task_name="Gaps"))

visualize_env(dmcenv.DMCEnv(walker_name="Ant", task_name="Floor"))

visualize_env(dmcenv.DMCEnv(walker_name="Humanoid", task_name="Escape"))

visualize_env(dmcenv.DMCEnv(walker_name="Humanoid", task_name="Gaps"))
```
